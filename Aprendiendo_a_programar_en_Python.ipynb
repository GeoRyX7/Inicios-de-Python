Strings

```python
#saltar lineas 
print("hola\nmundo")

#tabulacion de texto
print("hola\tmundo")
print("Nombre:\tGerald\nApellido:\tCasañe\nEdad:\t19")
print("\thola mundo")

#Caracter de escape \"
print("Ella dijo: \"Hola, ¿cómo estás?\"")

#formateo
name = "Gerald"
age = 19
print("mi nombre es {} y mi edad es {}".format(name, age))
print("Mi nombre es %s y mi edad es %d" % (name, age))

#formateo avanzado con f-strings
print(f"Mi nombre es {name} y mi edad es {age}")

#Desempaquetado de caracteres
word = "Python"
a, b, c, d, e, f = word
print(a, b, c, d, e, f)
print(a)
print(b)
print(c)
print(d)
print(e)
print(f)

#Dividir una palabra
lenguage = "python"
lenguage_slice = lenguage[0:3]
print(lenguage_slice)
lenguage_slice2 = lenguage[3:6]
print(lenguage_slice2)
print(lenguage_slice + lenguage_slice2)
lenguage_slice3 = lenguage[:3]
print(lenguage_slice3)
lenguage_slice4 = lenguage[3]
print(lenguage_slice4)
lenguage_slice5 = lenguage[-3:]
print(lenguage_slice5)

#reverse 
reversed_language = lenguage[::-1]
print(reversed_language)

#Dividir una cadena
sentence = "aprendiendo a programar en Python"
words = sentence.split()
print(words)
print(words[0])
print(words[1])

#reverse split
reversed_sentence = sentence[::-1]
print(reversed_sentence)
reversed_words = reversed_sentence.split()
print(reversed_words)

#Funciones de cadena 
print(lenguage.capitalize())
print(lenguage.upper())
print(lenguage.count("o"))
print(lenguage.isnumeric())
print("1".isnumeric())
print(lenguage.lower())
print(lenguage.upper().isupper())
print(lenguage.startswith("py"))
print(lenguage.endswith("on"))
print(lenguage.startswith("Py"))




```
