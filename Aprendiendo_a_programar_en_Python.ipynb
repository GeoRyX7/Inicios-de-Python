Variables

```python
variable_string = "Mi nombre es Gerald"
variable2_string = "Estudio ing fisica"
variable3_int = 19
variable4_float = 3.14
variable5_bool = True
variable6_list = [1, 2, 3, 4, 5]
variable7_complex = 1 + 2j
#Imprimo las variables
print(variable_string)
print(variable2_string)
print(variable3_int)
print(variable4_float)
print(variable5_bool)
print(variable6_list)
print(variable7_complex)
print(variable_string, variable2_string, variable3_int, variable4_float, variable5_bool, variable6_list, variable7_complex)
#Tipos de datos
print(type(variable_string))
print(type(variable2_string))
print(type(variable3_int))
print(type(variable4_float))
print(type(variable5_bool))
print(type(variable6_list))
print(type(variable7_complex))
#Transformacion de tipos de datos
#De string a int
Nueva_variable_string = str(variable3_int)
print(Nueva_variable_string)
print(type(Nueva_variable_string))
#De int a float
Nueva_variable_int = float(variable3_int)
print(Nueva_variable_int)
print(type(Nueva_variable_int))
#De float a int
Nueva_variable_float = int(variable4_float)
print(Nueva_variable_float)
print(type(Nueva_variable_float))
#De int a string
Nueva_variable_int_a_string = str(variable3_int)
print(Nueva_variable_int_a_string)
print(type(Nueva_variable_int_a_string))

```
